```python
import json

# similar for other data
#with open("LiveBench-ReasonFlux.json", "r", encoding="utf-8") as f:
with open("CodeContests_train.json", "r", encoding="utf-8") as f:
    data = json.load(f)

final_data = []

for i in range(len(data)):
    final_data_i = {}
    final_data_i["question"] = data[i]["question"]
    final_data_i["test_method"] = "stdio"
    final_data_i["test_time_limit"] = 6 # for eval
    #final_data_i["test_time_limit"] = 2 # for train
    final_data_i["test_input"] = data[i]["test_input"][:8]
    final_data_i["test_output"] = data[i]["test_output"][:8]
    final_data.append(final_data_i)


#with open("LiveBench-ReasonFlux.json", "w", encoding="utf-8") as f:
with open("CodeContests_train.json", "w", encoding="utf-8") as f:
    json.dump(final_data, f, indent=2, ensure_ascii=False)

len(final_data)
```
